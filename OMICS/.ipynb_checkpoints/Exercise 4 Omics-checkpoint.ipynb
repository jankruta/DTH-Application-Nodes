### Import libraries. Entrez is used to enter the ncbi database.

```python
from Bio import SeqIO
from Bio import Entrez
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
plt.ion()

import logomaker as lm
Entrez.email = "amun.stalder@students.fhnw.ch"
```

### Downloading the sequence of the human Myosine heavy chain 7, 8, 9, 3

```python
with Entrez.efetch(db = "nucleotide", rettype = "gb", retmode = "text", id = "NG_007884") as handle:
    for record in SeqIO.parse(handle, "gb"):
        MYH7 = str(record.seq)
        MYH7_50 = MYH7[0:50]
#downloading the sequence of the human Myosine heavy chain 8
with Entrez.efetch(db = "nucleotide", rettype = "gb", retmode = "text", id = "NG_013015") as handle:
    for record in SeqIO.parse(handle, "gb"):
        MYH8 = str(record.seq)
        MYH8_50 = MYH8[0:50]
#downloading the sequence of the human Myosine heavy chain 9
with Entrez.efetch(db = "nucleotide", rettype = "gb", retmode = "text", id = "NG_011884") as handle:
    for record in SeqIO.parse(handle, "gb"):
        MYH9 = str(record.seq)
        MYH9_50 = MYH9[0:50]
#downloading the sequence of the human Myosine heavy chain 3
with Entrez.efetch(db = "nucleotide", rettype = "gb", retmode = "text", id = "NG_011537") as handle:
    for record in SeqIO.parse(handle, "gb"):
        MYH3 = str(record.seq)
        MYH3_50 = MYH3[0:50] 
```

### Printing the first 50 nucleotides 

```python
print("Myosin heavy chain 3: " + MYH3_50)
print("Myosin heavy chain 7: " + MYH7_50)
print("Myosin heavy chain 8: " + MYH8_50)
print("Myosin heavy chain 9: " + MYH9_50)
```

### Making a nice Position Weight Matrix

```python
seqs = [MYH3_50, MYH7_50, MYH8_50, MYH9_50]
counts_mat = lm.alignment_to_matrix(seqs)
counts_mat.head()
```

### Making a nice Position Weight Logo

```python
lm.Logo(counts_mat)
```
